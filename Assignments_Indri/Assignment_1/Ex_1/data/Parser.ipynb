```python
import pandas as pd
```

```python
df = pd.read_csv("Output.dat", sep="\t", header=None)
```

```python
result = []
for i in range(20):
    result.append(df.iloc[i::20].sum().tolist())
```

```python
df = pd.DataFrame(result)
df.to_csv("ParsedOutputScaling.dat", sep="\t", header=False)
df
```

```python
# Speedup
df[df.columns[1]] = df[df.columns[0]]/df[df.columns[1]]
df[df.columns[2]] = df[df.columns[0]]/df[df.columns[2]]
df[df.columns[3]] = df[df.columns[0]]/df[df.columns[3]]
df
```

```python
df.to_csv("ParsedOutputSpeedup.dat", sep="\t", header=False)
```
